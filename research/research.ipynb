```python
example = ConfigBox({"key1":"value1","key2":"value2"})
example.key1
```

***ENSURE ANNOTATIONS*** if we have a function which accepts integer as input and gives integer as output and by mistake we give a integeer as string to it then normally it gets accepted and we get very wrong results but by asuinf ensure annotations it will raise error if something wrong is given

```python
def get_product(x, y):
    return x * y
```

```python
get_product(3, 4)
```

```python
get_product(3, "4")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product(x: int, y: int) -> int:
    return x * y
```

```python
get_product(4, "5")
```
